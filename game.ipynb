```python
# Filtrando pessoas com mais de 30 anos
filtro_ano = df['Year_of_Release'] > 2006
print(df[filtro_ano])
```

```python
# Filtrando pessoas que moram em São Paulo
filtro_pub = df['Developer'] == 'Nintendo'
print(df[filtro_pub])
```

```python
# Combinando filtros (idade > 30 E cidade != 'São Paulo')
print(df[(df['Year_of_Release'] >= 2010) & (df['Developer'] == 'Nintendo')])
```

```python
# Ordenando pela coluna 'Idade' em ordem crescente
df.sort_values(by='Year_of_Release')
```

```python
# Ordenando pela coluna 'Idade' em ordem decrescente
df.sort_values(by='User_Score', ascending=True)
```
